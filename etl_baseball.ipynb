```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "data/Batting.csv"
batting_df = pd.read_csv(csv_file)
batting_df.head()
```

```python
batting_df2 = batting_df[(batting_df['yearID'] == 2015)]
batting_df2.head()
```

```python
batting_df3 = batting_df2[(batting_df2['stint'] == 1)]
batting_df3.head()
```

```python
new_batting_df4 = batting_df3[['playerID', 'teamID', 'AB', 'R', 'HR', 'RBI', 'SB']].copy()
new_batting_df4.head()
```

```python
new_batting_df = new_batting_df4.rename(columns={"playerID": "playerid", "teamID": "teamid", "AB": "ab", "R": "r", "HR": "hr", "RBI": "rbi", "SB": "sb"})
```

```python
new_batting_df.head()
```

```python
csv_file2 = "data/Salaries.csv"
salaries_df = pd.read_csv(csv_file2)
salaries_df.head()
```

```python
salaries_df2 = salaries_df[(salaries_df['yearID'] == 2015)]
salaries_df2.head()
```

```python
new_salaries_df2 = salaries_df2[['playerID', 'salary']].copy()
new_salaries_df2.head()
```

```python
new_salaries_df = new_salaries_df2.rename(columns={"playerID": "playerid"})
```

```python
new_salaries_df.head()
```

```python
csv_file3 = "data/People.csv"
people_df = pd.read_csv(csv_file3)
people_df.head()
```

```python
new_people_df2 = people_df[['playerID', 'birthYear', 'birthCountry', 'nameFirst', 'nameLast']].copy()
new_people_df2.head()
```

```python
new_people_df = new_people_df2.rename(columns={"playerID": "playerid", "birthYear": "birthyear", "birthCountry": "birthcountry", "nameFirst": "namefirst", "nameLast": "namelast"})
```

```python
new_people_df.head()
```

```python
csv_file4 = "data/CollegePlaying.csv"
college_df = pd.read_csv(csv_file4)
college_df.head()
```

```python
college_df.drop_duplicates('playerID', inplace=True)
```

```python
college_df.head()
```

```python
college_df.reset_index(inplace=True)
```

```python
college_df.head()
```

```python
new_college_df2 = college_df[['playerID', 'schoolID']].copy()
new_college_df2.head()
```

```python
new_college_df = new_college_df2.rename(columns={"playerID": "playerid", "schoolID": "collegeid"})
```

```python
new_college_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/baseball"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_batting_df.to_sql(name='batting', con=engine, if_exists='append', index=False)
```

```python
new_salaries_df.to_sql(name='salaries', con=engine, if_exists='append', index=False)
```

```python
new_people_df.to_sql(name='people', con=engine, if_exists='append', index=False)
```

```python
new_college_df.to_sql(name='college', con=engine, if_exists='append', index=False)
```
